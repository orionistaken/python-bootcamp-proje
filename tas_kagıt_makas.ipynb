```python
import random

def oyun_aciklamalari():
    # Oyun açıklamaları
    print("Taş, kağıt, makas, su ve ateş oyununa hoşgeldin.")
    print("Taş makası ve ateşi yener")
    print("Kağıt taşı ve suyu yener")
    print("Makas kağıdı ve suyu yener")
    print("Su taşı ve ateşi yener")
    print("Ateş kağıdı ve makası yener")
    print("İlk önce 2 raunt kazanan oyunu kazanır!")
    print("Oyuna başlamak için taş, kağıt, makas, su veya ateşten birini seç çıkmak için 'q' yaz")

def oyuncu_secenek_girdisi(secenekler):
    while True:
        oyuncu_secimi = input("Taş, Kağıt, Makas, Su, Ateş? (Çıkmak için q'ya basın):").lower()
        if oyuncu_secimi == 'q':
            return oyuncu_secimi
        if oyuncu_secimi not in secenekler:
            print("Geçersiz bir seçenek, lütfen tekrar deneyin.")
        else:
            return oyuncu_secimi

def kazanani_sec(oyuncu_secimi, bilgisayar_secimi):
    # Kazanma koşulları
    if (oyuncu_secimi == 'taş' and (bilgisayar_secimi == 'makas' or bilgisayar_secimi == 'ateş') or
        oyuncu_secimi == 'kağıt' and (bilgisayar_secimi == 'taş' or bilgisayar_secimi == 'su') or
        oyuncu_secimi == 'makas' and (bilgisayar_secimi == 'kağıt' or bilgisayar_secimi == 'su') or
        oyuncu_secimi == 'su' and (bilgisayar_secimi == 'taş' or bilgisayar_secimi == 'ateş') or
        oyuncu_secimi == 'ateş' and (bilgisayar_secimi == 'kağıt' or bilgisayar_secimi == 'makas')):
        return 'oyuncu'
    elif oyuncu_secimi == bilgisayar_secimi:
        return 'berabere'
    else:
        return 'bilgisayar'        

def oyuna_devam_girdisi():
    while True:
        devam = input("Başka bir oyun oynamak ister misiniz? (evet/hayır): ").lower()
        if devam in ['evet', 'hayır']:
            return devam
        print("Lütfen 'evet' veya 'hayır' olarak yanıt verin.")    
    
def tas_kagit_makas_BURHAN_KORKMAZ():
    oyun_aciklamalari()

    # Oyunda kullanılabilecek seçenekler
    secenekler = ['taş', 'kağıt', 'makas', 'su', 'ateş']
    oyun_sayisi = 0
    
    while True:
        oyuncu_galibiyet = 0
        bilgisayar_galibiyet = 0
        oyun_sayisi += 1
        raunt_sayisi = 1

        print("\n###########################################")
        print(f"{oyun_sayisi}. Oyun Başlıyor!")
        print("###########################################\n")
        
        while oyuncu_galibiyet < 2 and bilgisayar_galibiyet < 2:
            print(f"#### {oyun_sayisi}. Oyun {raunt_sayisi}. Raunt başlıyor ####\n")
            oyuncu_secimi = oyuncu_secenek_girdisi(secenekler)
            
            if oyuncu_secimi == 'q' and oyun_sayisi == 1:
                print("Madem oynamayacaktın niye başlattın! Korktun mu?")
                return
            if oyuncu_secimi == 'q':
                print("Oyundan çıkılıyor... Oynadığın için teşekkürler.")
                return
            
            bilgisayar_secimi = random.choice(secenekler)
            print(f"Bilgisayarın seçimi: {bilgisayar_secimi}\n")

            sonuc = kazanani_sec(oyuncu_secimi, bilgisayar_secimi)
            
            if sonuc == 'oyuncu':
                print("Bu turu kazandınız!\n")
                print("Bilgisayar: Ahh şanslıydın.\n")
                oyuncu_galibiyet += 1
            elif sonuc == 'bilgisayar':
                print("Bilgisayar Kazandı")
                print("Bilgisayar: Hahahah kolaydı.\n")
                bilgisayar_galibiyet += 1
            else:
                print("Berabere, bu tur kazanan yok!!!\n")
                
            raunt_sayisi += 1
            print(f"Durum: Oyuncu {oyuncu_galibiyet} - Bilgisayar {bilgisayar_galibiyet}\n")
        
        if oyuncu_galibiyet == 2:
            print("Tebrikler! Oyunu kazandınız!")
        else:
            print("Bilgisayar oyunu kazandı!")
        
        devam = oyuna_devam_girdisi()
        bilgisayar_devam = random.choice(['evet', 'hayır'])
        print(f"Bilgisayarın oyuna devam etme isteği: {bilgisayar_devam}")
        
        if devam != 'evet' or bilgisayar_devam != 'evet':
            print("Oyun sonlandırılıyor. Görüşmek üzere!")
            break
        else:
            print("Hadi bir oyun daha oynayalım.")
```

```python
tas_kagit_makas_BURHAN_KORKMAZ()
```
